# Plotting a Simple Line Graph

```python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

# Using predefined styles in matplotlib pyplot
plt.style.use('Solarize_Light2') 
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=3)

# Set chart title and label axes
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels
ax.tick_params(axis='both', labelsize=14)
plt.show()
```

# Scatter plots

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots()
ax.scatter(2, 4, s=200)

# Set chart title and label axes
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels
ax.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

Plotting a series of points

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100)

# Set the chart title and label axes
ax.set_title("Square of numbers", fontsize=14)
ax.set_ylabel("Squares", fontsize=14)
ax.set_xlabel("Numbers", fontsize=14)

ax.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

Using list comprehensions

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots()

# The use of c specifies which axis to use and cmap which colormap
ax.scatter(x_values, y_values, s=10, c=y_values, cmap=plt.cm.Blues)

ax.set_title("Square of numbers", fontsize=14)
ax.set_ylabel("Squares", fontsize=14)
ax.set_xlabel("Numbers", fontsize=14)

# Set the range for each axes
ax.axis([0, 1100, 0, 1100000])

plt.show()
```

Plotting 5 cubed numbers

```python
x_values = [1, 2, 3, 4, 5]
y_values = [x**3 for x in x_values]

plt.style.use('seaborn-v0_8-paper')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=50)

ax.set_title("Cubed Numbers", fontsize=20)
ax.set_ylabel("Cubes", fontsize=20)
ax.set_xlabel("Numbers", fontsize=20)

ax.tick_params(axis='both', labelsize=10)
plt.show()
```

Plotting 5000 cubed numbers

```python
x_values = range(1, 5001)
y_values = [x**3 for x in x_values]

# Pick the styling
plt.style.use('seaborn-v0_8-paper')
fig, ax = plt.subplots()

# Using colormap
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Reds, s=10)
ax.set_title("Cubed Numbers", fontsize=20)
ax.set_ylabel("Cubes", fontsize=20)
ax.set_xlabel("Numbers", fontsize=20)

plt.show()
```
