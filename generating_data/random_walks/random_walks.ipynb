# Random Walks

```python
import matplotlib.pyplot as plt
from random_walk import RandomWalk

# Make a random walk.
rw = RandomWalk(500_000)
rw.fill_walk()

# Plot the points in the walk
plt.style.use('classic')
fig, ax = plt.subplots(figsize=(15, 9))
ax.scatter(
    rw.x_values,
    rw.y_values,
    c=range(rw.num_points),
    s=1,
    cmap=plt.cm.Blues,
    edgecolors='none'
)

# Remove the axes to make it cleaner
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

# Emphasize the first and last points (Doesn't look as nice)
# ax.scatter(0, 0, c='green', edgecolors='none', s=30)
# ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=30)

plt.show()
```

Using plot instead of scatter

```python
rw = RandomWalk()
rw.fill_walk()

fig, ax = plt.subplots(figsize=(15, 9))
ax.plot(rw.x_values, rw.y_values, c='Blue', linewidth=1)

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.show()
```
