# Using both matplotlib and plotly

Matplotlib to plot dice rolls

```python
import matplotlib.pyplot as plt
from die import Die

# List to hold all the dies
dies = [Die(i) for i in range(6, 10)]

# Store max and min values
max_val = 0
min_val = len(dies)
for die in dies:
    max_val += die.num_sides

# Roll the dice and store their results
results = []
for roll_num in range(10_000):    
    total_roll = 0
    for die in dies:
        total_roll += die.roll()
    results.append(total_roll)
    
# Compute the frequencies of each result
frequencies = []
for result in range(min_val, max_val+1):
    frequencies.append(results.count(result))

# Now we create the plot
plt.style.use('classic')
x_values = range(min_val, max_val+1)

fig, ax = plt.subplots(figsize=(15, 9))
ax.bar(x=x_values, height=frequencies)
ax.set_title('Results of rolling D6, D7, D8 and D9 dice')
ax.set_xlabel('Result', fontsize=14)
ax.set_ylabel('Frequency of Result', fontsize=14)

plt.show()
```
